```python
import os
import sys
import glob
import matplotlib.pyplot as plt
import matplotlib.patches as patch
import numpy as np
import pandas as pd
%matplotlib inline
%precision 4
plt.style.use('ggplot')
from scipy import linalg
np.set_printoptions(suppress=True)

# Students may (probably should) ignore this code. It is just here to make pretty arrows.

def plot_vectors(vs):
    """Plot vectors in vs assuming origin at (0,0)."""
    n = len(vs)
    X, Y = np.zeros((n, 2))
    U, V = np.vstack(vs).T
    plt.quiver(X, Y, U, V, range(n), angles='xy', scale_units='xy', scale=1)
    xmin, xmax = np.min([U, X]), np.max([U, X])
    ymin, ymax = np.min([V, Y]), np.max([V, Y])
    xrng = xmax - xmin
    yrng = ymax - ymin
    xmin -= 0.05*xrng
    xmax += 0.05*xrng
    ymin -= 0.05*yrng
    ymax += 0.05*yrng
    plt.axis([xmin, xmax, ymin, ymax])
```

```python
a1 = np.array([7,0])         #矩陣a1
a2 = np.array([0,5])         #矩陣a2
A = np.array([[2,1],[1,1]])  #矩陣A
v2 =np.dot(A,v1)             #v2為A跟v1進行矩陣乘法
plt.figure(figsize=(8,8))    #做出來的圖size為8*8
plot_vectors([a1, a2])       #將矩陣a1,a2相加成一個矩陣並用像量形式作圖
v1 = np.array([2,3])         #v1向量
plot_vectors([v1,v2])        #畫出向量v1,v2
plt.text(2,3,"v1 =(2,3)",fontsize=16)     #在位置(2,3)加入"v1=(2,3)"的文字
plt.text(6,5,"Av1 = ", fontsize=16)       #在位置(6,5)加入"Av1 = "的文字
plt.text(v2[0],v2[1],"(7,5)",fontsize=16) #在位置(7,5)加入"(7,5)"的文字
print(v2[0],v2[1])                        #印出v2[0],v2[1]的值
```
